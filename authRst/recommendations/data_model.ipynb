```python

num_values = 10000
prob_success = 0.5 
values = np.random.binomial(n=90-18+1, p=prob_success, size=num_values) + 18

df_age = pd.DataFrame({'idade': values})
```

```python
df = pd.DataFrame(create_dataset(10000,900))
```

```python
df = pd.concat([df, df_age], axis=1)
```

```python
df
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_data, columns=df.columns)
```

```python
array_data = scaled_df.values
```

```python
kmeans_model = KMeans(n_clusters=4)
```

```python
clusters = pd.DataFrame(kmeans_model.fit_predict(array_data))
```

```python
# Coverte o array de volta a df
columns = df.columns
array_data = pd.DataFrame(array_data, columns=columns)
```

```python
array_data.insert(array_data.columns.get_loc("acompanhamento"), "Cluster", clusters)
df.insert(df.columns.get_loc("acompanhamento"), "Cluster", clusters)
array_data.head()

```
